```python
import pandas as pd
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
from google.colab import drive
drive.mount('/content/drive')

df=pd.read_csv('/content/drive/MyDrive/Social_Network_Ads (1).csv')
df.head(5)
```

```python
df.isnull().sum()
```

```python
df_columns = df.columns
```

```python
pred = df[df_columns[df_columns!="Age"]]
targ = df["Age"]
```

```python
targ.head()
```

```python
pred_normal = ((pred - pred.mean())/pred.std())
pred_normal.head()
```

```python
n_cols = pred_normal.shape[1]
```

```python
def regression_model():
    model = Sequential()
    model.add(Dense(50,activation = "relu",input_shape = (n_cols,)))
    model.add(Dense(50,activation ="relu"))
    model.add(Dense(50,activation ="relu"))
    model.compile(optimizer = "adam", loss = "mean_squared_error")

    return model
```

```python
model = regression_model()
model.fit(pred_normal,targ,validation_split = 0.3, epochs = 100, verbose = 1)
```
